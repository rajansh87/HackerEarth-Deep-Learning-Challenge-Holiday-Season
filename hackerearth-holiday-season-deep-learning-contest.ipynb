# Model Creation

```python
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
#inputlayer : apply filters
model.add(Convolution2D(filters=32, 
                        kernel_size=(3,3), 
                        activation='relu',
                   input_shape=(64, 64, 3)
                       ))
```

```python
# pooling layer where we are doing maxpooling
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
#modification for increasing accuracy
model.add(Convolution2D(filters=32, 
                        kernel_size=(3,3), 
                        activation='relu',
                       ))
```

```python
#modification for increasing accuracy
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
#layer inwhich we areconverting 2d/3d image to 1d image i.e flattening
model.add(Flatten())
```

```python
# from here our hidden layerrs starts
model.add(Dense(units=128, activation='relu'))
```

```python
#output layer : to provide binary output using sigmoid function
model.add(Dense(units=6, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

# Image Preprocessing

```python
from keras_preprocessing.image import ImageDataGenerator
```

```python
#image augmentation
#url : https://keras.io/api/preprocessing/image/ 
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
training_set = train_datagen.flow_from_directory(
        './train_data/',
        target_size=(64,64),
        batch_size=32,
        class_mode='categorical')
test_set = test_datagen.flow_from_directory(
        '../input/hackereath-holiday-season-deep-learning-contest/dataset/test/',
        target_size=(64,64),
        batch_size=32,
        class_mode='categorical')
```

```python
training_set.class_indices # to see classes of our dataset
```

# Model Training

```python
history=model.fit(
        training_set,
        steps_per_epoch=6469//32,
        epochs=50,
        validation_data=test_set,
        validation_steps=3489//32)
```

# Final Prediction CSV file

```python
def final_Output(result):
    if result[0][0]==1:
        return("Airplane")
    elif result[0][1]==1:
        return("Candle")
    elif result[0][2]==1:
        return("Christmas_Tree")
    elif result[0][3]==1:
        return("Jacket")
    elif result[0][4]==1:
        return("Miscellaneous")
    else:
        return("Snowman")
```

```python
predictions=[]
images=[]

test_path='../input/hackereath-holiday-season-deep-learning-contest/dataset/test/'
for img in os.listdir(test_path):
    test_image = image.load_img(test_path+img,target_size=(64,64))    
    test_image = image.img_to_array(test_image)  #convert PIL image to numpy array
    test_image = np.expand_dims(test_image,axis=0)
    #since keras uses tensor flow and for tensorflow it needs 4d image so we converted 3d image to 4d image using above
    result = model.predict(test_image)
    pred=final_Output(result)
    predictions.append(pred)
    images.append(img)
final_answers=pd.DataFrame({'Image':images,'Class':predictions})
final_answers.to_csv('predictions.csv',index=False)
```
